```python
# Load the dataset
unemp = pd.read_csv("C:/Users/SAI CHARAN/Desktop/Unemployment_rates.csv")
unemp.head()
```

```python
unemp.info()
```

```python
# changing date datatype to datetime

unemp['TimePeriod'] = pd.to_datetime(unemp['TimePeriod'])
unemp.describe()
```

```python
plt.style.use('fivethirtyeight')
plt.figure(figsize=(15,7))
plt.plot(unemp['TimePeriod'], unemp['Total'])
```

```python
# unemployment rate for women
plt.style.use('fivethirtyeight')
plt.figure(figsize=(15,7))
sns.lineplot(data=unemp, x='TimePeriod', y='Female', color='blue')

plt.title('Unemployment rate for Women',fontdict= { 'fontsize': 24, 'fontweight':'bold'})
plt.ylabel('Percent', fontdict= { 'fontsize': 15, 'fontweight':'bold'})
plt.xlabel('Date', fontdict= { 'fontsize': 15, 'fontweight':'bold'})


plt.show()
```

```python
# unemployment rate for men
plt.style.use('fivethirtyeight')
plt.figure(figsize=(15,7))
sns.lineplot(data=unemp, x='TimePeriod', y='Male', color='red')

plt.title('Unemployment rate for Men',fontdict= { 'fontsize': 24, 'fontweight':'bold'})
plt.ylabel('Percent', fontdict= { 'fontsize': 15, 'fontweight':'bold'})
plt.xlabel('Date', fontdict= { 'fontsize': 15, 'fontweight':'bold'})


plt.show()
```

```python
#Unemployment rate by age between 2010 to 2020
plt.style.use('fivethirtyeight')
plt.figure(figsize=(15,10))
sns.lineplot(data=unemp[(unemp.TimePeriod >= '2010-03-31')&(unemp.TimePeriod<='2020-12-31')],x="TimePeriod", y="Aged 15-24 Years", label="Aged 15-24 Years")
sns.lineplot(data=unemp[(unemp.TimePeriod >= '2010-03-31')&(unemp.TimePeriod<='2020-12-31')],x="TimePeriod", y="Aged 25-34 Years", label="Aged 25-34 Years")
sns.lineplot(data=unemp[(unemp.TimePeriod >= '2010-03-31')&(unemp.TimePeriod<='2020-12-31')],x="TimePeriod", y="Aged 35-44 Years", label="Aged 35-44 Years")
sns.lineplot(data=unemp[(unemp.TimePeriod >= '2010-03-31')&(unemp.TimePeriod<='2020-12-31')],x="TimePeriod", y="Aged 45-54 Years", label="Aged 45-54 Years")
sns.lineplot(data=unemp[(unemp.TimePeriod >= '2010-03-31')&(unemp.TimePeriod<='2020-12-31')],x="TimePeriod", y="Aged 55-64 Years", label="Aged 55-64 Years")
sns.lineplot(data=unemp[(unemp.TimePeriod >= '2010-03-31')&(unemp.TimePeriod<='2020-12-31')],x="TimePeriod", y="Aged 65 Years and Over", label="Aged 65 Years and Over")

plt.title('Unemployment rate by Age',fontdict= { 'fontsize': 24, 'fontweight':'bold'})
plt.ylabel('Percent', fontdict= { 'fontsize': 15, 'fontweight':'bold'})
plt.xlabel('Date', fontdict= { 'fontsize': 15, 'fontweight':'bold'})


plt.show()
```

```python
#Unemployment rate by Gender between 2010 to 2020

plt.style.use('fivethirtyeight')
plt.figure(figsize=(15,7))
sns.lineplot(data=unemp[(unemp.TimePeriod >= '2010-03-31')&(unemp.TimePeriod<='2020-12-31')],x="TimePeriod", y="Female",label="Female")
sns.lineplot(data=unemp[(unemp.TimePeriod >= '2010-03-31')&(unemp.TimePeriod<='2020-12-31')],x="TimePeriod", y="Male",label="Male")

plt.title('Unemployment rate by Gender',fontdict= { 'fontsize': 24, 'fontweight':'bold'})
plt.ylabel('Percent', fontdict= { 'fontsize': 15, 'fontweight':'bold'})
plt.xlabel('Date', fontdict= { 'fontsize': 15, 'fontweight':'bold'})


plt.show()
```

```python
# Analysing unemployment during COVID

covid = unemp[(unemp.TimePeriod >= '2019-12-31')]
covid.tail()
```

```python
plt.figure(figsize=(15,7))
sns.lineplot(data = covid, x='TimePeriod', y='Total')
```

```python
# Unemployement rate during COVID by Gender

plt.figure(figsize=(15,7))
sns.lineplot(data = covid, x='TimePeriod', y='Male', label='Men', color='Blue')
sns.lineplot(data = covid, x='TimePeriod', y='Female', label='Women', color='brown');
plt.legend

plt.title('Unemployement rate during COVID by Gender',fontdict= { 'fontsize': 24, 'fontweight':'bold'})
plt.ylabel('Percent', fontdict= { 'fontsize': 15, 'fontweight':'bold'})
plt.xlabel('Date', fontdict= { 'fontsize': 15, 'fontweight':'bold'})


plt.show()
```

```python
# Unemployement rate during COVID by Race

plt.figure(figsize=(15,7))
sns.lineplot(data = covid, x='TimePeriod', y='Asian', label='Asian')
sns.lineplot(data = covid, x='TimePeriod', y='European', label='European')
sns.lineplot(data = covid, x='TimePeriod', y='Maori', label='Maori')
sns.lineplot(data = covid, x='TimePeriod', y='Pacific Peoples', label='Pacific Peoples');
plt.legend

plt.title('Unemployement rate during COVID by Race',fontdict= { 'fontsize': 24, 'fontweight':'bold'})
plt.ylabel('Percent', fontdict= { 'fontsize': 15, 'fontweight':'bold'})
plt.xlabel('Date', fontdict= { 'fontsize': 15, 'fontweight':'bold'})


plt.show()
```
